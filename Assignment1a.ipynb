# 1. <code>arange</code> and <code>linspace</code>

First, let's do a little review.  Recall that the <code>arange</code> function in numpy can be used to generate numpy arrays with a sequence of numbers.  The code below shows how to use <code>arange</code> to create a list with 100 equally spaced elements, starting at 0 and ending on 0.99.  Just to check this, we will plot it, and we will print out the first and last element to verify that they are indeed 0 and 0.99, respectively

```python
x = np.arange(100) / 100
plt.plot(x)
plt.title("np.arange(100)/100")
```

```python
print(x[0])
print(x[-1])
```

What's slightly more difficult, but doable with some thought, is to create 100 equally spaced elements starting at 0 and ending at <b>1</b>.  If you think about it for a moment, you'll realize that the interval length between each element should be $1/99$, which is slightly longer than the $1/100$ we just used.  

```python
interval = 1.0/99.0
x = interval*np.arange(100)
plt.plot(x)
plt.title("np.arange(100)*(1/99)")
```

```python
print(x[0]) # The first element is still 0
print(x[-1]) # But the last element is now 1
```

What if we want to come up with a list of 32 evenly spaced numbers starting at $-\pi$ and ending at $\pi$? (Note that $\pi$ in numpy is <code>np.pi</code>).  Well, this is starting to get a little old to do manually.  Instead, numpy has a function designed for this called <code>linspace</code>.  You can type <code>linspace?</code> for more details, but basically, it takes three parameters: <ol><li>The starting number</li><li>The ending number</li><li>The total length of the list</li></ol>.  So to do what we said above, the command would be

```python
x = np.linspace(-np.pi, np.pi, 32)
```

Just to mix it up, let's plot it with a stem plot

```python
plt.stem(x) 
```

One neat thing we can use this for is to sample a function on some interval of our choosing.  So, for instance, if we want to plot $f(x) = x^2 + 5$ on the interval, $x \in [-3, 3]$, using 100 samples, we would write the following code

```python
x = np.linspace(-3, 3, 100)
y = x**2 + 5  # We do an element-wise square and an element-wise addition of 5
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("$f(x) = x^2 + 5$")
```

Notice how when we call <code>plot</code> with two lists, the first list is the values of $x$, and the second list is the values of $y$.  Both of these lists must be the same length, and their elements must be in direct correspondence.  So, for instance, the ordered pair $(x[10], y[10])$ will be on the graph

## 2: Sinusoids
The sine function, written as $f(x) = \sin(x)$, is one of the fundamental <i>trigonometric functions</i> which is defined using a right triangle.  More specifically, given an angle $x$ in radians, the sine gives the ratio of the side of a right triangle opposite x to the length of its hypotenuse.  Let's plot it over the interval $[0, 2 \pi]$ to see its remarkable shape

```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x))

# Don't worry about this line too much...it's just
# for making pretty ticks on the x-axis
plt.xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi], ["0", "$\pi/2$", "$\pi$", "$3 \pi/2$", "$2 \pi$"])
```

The sine function is an example of a <i>periodic function</i>; that is, it <b>repeats itself</b> on every interval of $2 \pi$.  To see this, let's plot it over the interval from $-4 \pi$ to $4 \pi$

```python
x = np.linspace(-4*np.pi, 4*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y)

# Again, don't worry too much about the line below
plt.xticks([np.pi*i for i in range(-4, 5)], ["$%i \pi$"%i for i in range(-4, 5)])
```

Therefore, the sine function is the basic building block of all <i>oscillatory</i> (repeating) phenomena in nature.

To make different versions of the sine that either oscillate more rapidly or more slowly, we can use the following equation
<h2>$f(t) = \sin(2 \pi f t)$</h2>

Where $f$ is the <i>frequency</i> of oscillation, and it specifies how many oscillations the sine should go through on an interval of length 1.  Note that we have switched the name of the independent variable from $x$ to $t$, because the independent variable is often time in this context, and this is what it will be for our audio applications as well.  

The <i>period</i> of oscillation, usually referred to as $T$, is $T = 1/f$.  If the units of $t$ and $T$ are seconds, then the units of $f$ are referred to as <b>herz (hz)</b>, or cycles/second.
